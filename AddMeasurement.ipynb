```
import pandas as pd
```

```
# Storm price 27.58
data_file = 'drive/My Drive/Expenses/measurements.csv'
data = pd.read_csv(data_file)
data
```

```
 while True:
    answer = input("Add new measurement (y/n): ")
    if answer != "y": 
      break;

    measurement_type = input("Enter 0 for Storm 1 for Gas: ")
    while measurement_type != "0" and measurement_type != "1":
      measurement_type = input("Enter 0 for EUR 1 for MKD:")
    if measurement_type == '0':
      measurement_t = 'Storm'
      measure = 'kWh'
      unit_price = 0.23
    elif measurement_type == '1':
      measurement_t = 'Gas'
      measure = 'mWh'
      unit_price = 63.2

    currency = 'EUR'
    measurement: float = input("Enter the measurement: ")

    is_now = input("Is the measurement taken now? (y/n): ")
    while is_now != "y" and is_now != "n":
      is_now = input("Is the measurement taken now? (y/n): ")

    if is_now == 'y':
      date = pd.Timestamp.now()
    else:
      date = pd.Timestamp(input("Enter the time when the measurement is taken: "))


    row = {'measure_unit': measure, 'type': measurement_t, 'aggregate_consumption': measurement, 'date': date, 'currency': currency, 'unit_price': unit_price}
    data = data.append(row, ignore_index=True, sort=False)
    print(row)
        
data.to_csv(data_file, index=False)

```
