```
import pandas as pd
```

```
data_file = 'drive/My Drive/Expenses/expenseList.csv'
data = pd.read_csv(data_file)
data
```

```
while True:
    answer = input("Add new expense:(y/n)")
    if answer == "y":
        amount: float = input("Enter the amount of the expense:")
        currency_index = input("Enter 0 for EUR 1 for MKD:")
        while currency_index != "0" and currency_index != "1":
          currency_index = input("Enter 0 for EUR 1 for MKD:")
        if currency_index == '0':
          currency = 'EUR'
        elif currency_index == '1':
          currency = 'MKD'
        date = pd.Timestamp.now()
        payer = input("Enter who payed:(V/A)")
        while payer != "V" and payer != "A":
            payer = input("Enter who payed:(V/A)")
        data = data.append({'Currency': currency, 'Date': date, 'Amount': amount, 'Payer': payer}, ignore_index=True, sort=False)
        #data.loc[len(data)] = [amount, payer]
        print(data.head())
    else:
        break
        
data.to_csv(data_file, index=False)
```
