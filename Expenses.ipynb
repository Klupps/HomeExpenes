```
import pandas as pd
```

```
data_file = 'drive/My Drive/Expenses/expenseList.csv'
data = pd.read_csv(data_file)
data
```

```
total = data[['Amount']].sum(axis=0)['Amount']
print('Total:\t' + str(total))

grouped_data = data.groupby(['Payer']).sum()
acko_expenses = grouped_data.Amount.A
viki_expenses = grouped_data.Amount.V

print('Acko has spent ' + str(acko_expenses))
print('Viki has spent ' + str(viki_expenses))
print('\n')

if grouped_data.Amount.A < grouped_data.Amount.V:
    print('Acko needs to give ' + str(total / 2 - acko_expenses) + " to Viki")
elif grouped_data.Amount.A > grouped_data.Amount.V:
    print('Viki needs to give ' + str(total / 2 - viki_expenses) + " to Acko")
else:
    print('No one needs to give money')
```
